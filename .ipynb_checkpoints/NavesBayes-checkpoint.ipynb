```python
def probability_calc(mean,var_smooth,x_new):
    exp=-(x_new-mean)**2/(var_smooth*2)
    prob = (1 / np.sqrt(2 * np.pi * var_smooth)) * np.exp(exp)
    return prob
```

```python

alphas=[0.0001,0.001,0.01,0.1,1,10]
def train_models(X_train,Y_train):
    labels = np.unique(Y_train)
    var={}
    mean={}
    for alpha in alphas:
        mean[alpha] = {}
        var[alpha] = {}
        for label in labels:        
            X_label = X_train[Y_train == label]
            mean[alpha][label] = X_label.mean(axis=0)
            var[alpha][label] = X_label.var(axis=0) + alpha
    return var,mean
var,mean=train_models(x_train,y_train)
```

```python
def fit(Y_test, X_test,Y_train, var, mean, labels):
    accuracy = {}
    probs = {}
    for alpha in alphas:
        probs[alpha]={}
        accurate=0;
        for i, x_data in enumerate(X_test.values):
            class_probs = {}
            for label in labels:
                mu=mean[alpha][label]
                sig2=var[alpha][label]
                prob=np.log(np.sum(Y_train==label)/len(Y_train))
                prob+=np.sum(np.log(probability_calc(mu,sig2,x_data)+1e-9))
                class_probs[label]=prob
            predicted_label = max(class_probs, key=class_probs.get)
            if predicted_label == Y_test.values[i]:
                accurate += 1
            probs[alpha][i] = class_probs
        accuracy[alpha] = accurate / len(Y_test)
    return accuracy, probs

labels = np.unique(y_train)  
accuracy, probs = fit(y_test, x_test,y_train, var, mean, labels)
accuracy
```

```python
max_a = max(accuracy, key=accuracy.get)
max_a
```

```python
def predict(Y_test,X_test,Y_train,mean,var,labels,alpha):
    ans=[]
    for i, x_data in enumerate(X_test.values):
        class_probs = {}
        for label in labels:
            mu=mean[alpha][label]
            sig2=var[alpha][label]
            prob=np.log(np.sum(Y_train==label)/len(Y_train))
            prob+=np.sum(np.log(probability_calc(mu,sig2,x_data)+1e-9))
            class_probs[label]=prob
        predicted_label = max(class_probs, key=class_probs.get)
        ans.append(predicted_label)
    return ans
```

```python
# confusion matrix
def Confusion_matrix(y_test, x_test,y_train, mean, var,max_a):
    pred=predict(y_test, x_test,y_train, mean, var, labels,max_a)
    pred = np.array(pred).astype(int)
    original=np.array(y_test)
    TP=0
    TN=0
    FN=0
    FP=0
    for i in range (len(pred)):
        if(pred[i]==1 and original[i]==1):
            TP=TP+1
        if(pred[i]==0 and original[i]==0):
            TP=TP+1
        if(pred[i]==1 and original[i]==0):
            FP=FP+1
        if(pred[i]==0 and original[i]==1):
            FN=FN+1
    print(TP,' ',FP)
    print(FN,' ',TN)
```
